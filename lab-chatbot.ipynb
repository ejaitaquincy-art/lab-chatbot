```python
client = OpenAI(
    # This is the default and can be omitted
    api_key=OPENAI_API_KEY,
)

def get_completion(prompt, model="gpt-4o-mini", temperature=0):
    
    messages = [{"role": "user", "content": prompt}]
    response = client.chat.completions.create(
        model=model,
        messages=messages,
        temperature=temperature, 
    )
    return response.choices[0].message.content


def get_completion_from_messages(messages, model="gpt-4o-mini", temperature=0):
 
    response = client.chat.completions.create(
        model=model,
        messages=messages,
        temperature=temperature, 
    )
    return response.choices[0].message.content
```

# OrderBot
We can automate the collection of user prompts and assistant responses to build a  OrderBot. The OrderBot will take orders at a pizza restaurant. 

```python
def collect_messages(_):
    prompt = inp.value_input
    inp.value = ''
    context.append({'role':'user', 'content':f"{prompt}"})
    response = get_completion_from_messages(context) 
    context.append({'role':'assistant', 'content':f"{response}"})
    panels.append(
        pn.Row('User:', pn.pane.Markdown(prompt, width=600)))
    panels.append(
        pn.Row('Assistant:', pn.pane.Markdown(response, width=600, styles={'background-color': '#F6F6F6'})))
 
    return pn.Column(*panels)
```

```python
import panel as pn  # GUI
pn.extension()

panels = [] # collect display 

context = [ {'role':'system', 'content':"""
You are OrderBot, an automated service to collect orders for a pizza restaurant. \
You first greet the customer, then collects the order, \
and then asks if it's a pickup or delivery. \
You wait to collect the entire order, then summarize it and check for a final \
time if the customer wants to add anything else. \
If it's a delivery, you ask for an address. \
Finally you collect the payment.\
Make sure to clarify all options, extras and sizes to uniquely \
identify the item from the menu.\
You respond in a short, very conversational friendly style. \
The menu includes \
pepperoni pizza  12.95, 10.00, 7.00 \
cheese pizza   10.95, 9.25, 6.50 \
eggplant pizza   11.95, 9.75, 6.75 \
fries 4.50, 3.50 \
greek salad 7.25 \
Toppings: \
extra cheese 2.00, \
mushrooms 1.50 \
sausage 3.00 \
canadian bacon 3.50 \
AI sauce 1.50 \
peppers 1.00 \
Drinks: \
coke 3.00, 2.00, 1.00 \
sprite 3.00, 2.00, 1.00 \
bottled water 5.00 \
"""} ]  # accumulate messages


inp = pn.widgets.TextInput(value="Hi", placeholder='Enter text here')
button_conversation = pn.widgets.Button(name="Chat!")

interactive_conversation = pn.bind(collect_messages, button_conversation)

dashboard = pn.Column(
    inp,
    pn.Row(button_conversation),
    pn.panel(interactive_conversation, loading_indicator=True, height=300),
)

dashboard
```

## Prompt Experimentation Report  Personal Budget Analyzer

### Introduction

The objective of this experiment was to evaluate how prompt structure, specificity, and temperature settings influence the behavior and reliability of a large language model when analyzing structured financial data. The task consisted of building a Personal Budget Analyzer capable of extracting insights from a list of monthly expenses.

### Prompt Version 1  Structured JSON Request

The first prompt required the model to return a JSON object containing total expenses, largest category, non-essential expenses, and estimated savings. The instruction explicitly required valid JSON. 

The model successfully computed totals and identified spending categories correctly. The output was structured and machine-readable. However, minor formatting variations appeared depending on the phrasing of the instruction. This demonstrated that clear structural instructions significantly improve reliability.

### Prompt Version 2  Strict Constraint Enforcement

The second version introduced stronger constraints: strictly valid JSON, numeric-only monetary values, and no explanations. This reduced ambiguity and improved consistency. 

The output was more deterministic and compliant. The model adhered more strictly to formatting requirements. This confirms that increasing constraint precision enhances structured data extraction and reduces variability.

### Prompt Version 3  Advisory and Creative Mode

The third version shifted toward an advisory tone and used a higher temperature (0.7). Instead of strict JSON, the model provided qualitative financial analysis, risk assessment, and optimization suggestions.

While the insights were valuable and context-aware, the output was less structured and not ideal for automated processing. This version demonstrated that higher temperature increases creativity but reduces structural control.

### Key Findings

This experiment highlights several important principles of prompt engineering:

1. Specificity directly improves structural compliance.
2. Explicit formatting constraints reduce output variability.
3. Lower temperature increases determinism.
4. Higher temperature enhances creativity but reduces predictability.
5. Clear role definition influences response tone and analytical depth.

### Conclusion

Prompt design strongly affects output reliability and usability. For structured financial analysis tasks, strict instructions and low temperature settings produce the most consistent and automation-ready outputs. For advisory or interpretative tasks, a more flexible prompt and moderate temperature can generate richer qualitative insights. Understanding this balance is essential when designing AI-assisted analytical systems.
